## **Model & Tujuan**

**1. Clustering Menggunakan DBSCAN**

DBSCAN merupakan algoritma unsupervised learning yang digunakan untuk mengelompokkan provinsi di Indonesia berdasarkan kesamaan karakteristik demografi dan dampak COVID-19. Berbeda dengan metode clustering lain yang memaksa seluruh data masuk ke dalam suatu kelompok, DBSCAN bekerja berdasarkan tingkat kerapatan data sehingga mampu membedakan antara pola mayoritas dan data yang bersifat ekstrem.

Tujuan utama penggunaan DBSCAN adalah untuk mendeteksi anomali (outlier) serta mengidentifikasi pola dominan antar provinsi. Provinsi yang memiliki karakteristik sangat berbeda dari mayoritas nasional akan terdeteksi sebagai noise (label −1), sedangkan provinsi dengan karakteristik yang relatif serupa akan membentuk cluster inti.

Insight yang diperoleh:
* Identifikasi Wilayah Outlier: Provinsi yang terdeteksi sebagai noise menunjukkan karakteristik ekstrem, misalnya kepadatan penduduk yang sangat tinggi atau tingkat kematian yang jauh di atas rata-rata nasional. Wilayah seperti ini mengindikasikan bahwa kebijakan penanganan umum kurang efektif, sehingga diperlukan pendekatan yang lebih spesifik.
* Pola Mayoritas Provinsi: Cluster inti merepresentasikan kelompok provinsi dengan karakteristik yang relatif seragam. Untuk kelompok ini, pemerintah dapat menerapkan kebijakan yang lebih umum dan terstandarisasi.

**2. Clustering Menggunakan K-Means**

Berbeda dengan DBSCAN, K-Means mengharuskan seluruh provinsi masuk ke dalam salah satu cluster, sehingga hasil pengelompokan menjadi lebih terstruktur dan mudah diinterpretasikan.

Tujuan penggunaan K-Means adalah untuk membagi provinsi ke dalam beberapa tingkat risiko, seperti risiko rendah, sedang, dan tinggi, berdasarkan karakteristik dampak COVID-19 dan faktor demografi.

Insight yang diperoleh:
* Segmentasi Tingkat Risiko: K-Means membantu memetakan provinsi ke dalam kelompok risiko yang jelas, sehingga memudahkan perbandingan antar wilayah.
* Perbedaan Karakteristik Antar Cluster: Setiap cluster dapat dianalisis lebih lanjut untuk melihat perbedaan rata-rata jumlah kasus, tingkat kematian, kepadatan penduduk, dan tingkat urbanisasi, yang berguna untuk menentukan prioritas penanganan.

**3. Forecasting Time Series Menggunakan MLP**

Model Multi-Layer Perceptron (MLP) digunakan sebagai pendekatan deep learning untuk melakukan peramalan data time series COVID-19. Analisis difokuskan pada data nasional Indonesia atau provinsi kunci seperti DKI Jakarta dengan menggunakan data historis harian, seperti New Cases, New Active Cases, dan New Deaths.

Tujuan penggunaan MLP adalah untuk memprediksi tren pandemi dalam jangka pendek, yaitu sekitar 7–14 hari ke depan, dengan mempertimbangkan pola non-linear pada data.

Insight yang diperoleh:
* Deteksi Dini Lonjakan Kasus: Hasil prediksi MLP dapat digunakan sebagai indikator awal apakah pandemi berpotensi memasuki fase peningkatan atau penurunan kasus.

* Dukungan Pengambilan Kebijakan: Informasi hasil peramalan dapat menjadi bahan pertimbangan dalam menentukan kebijakan pembatasan sosial, seperti apakah perlu dilakukan pengetatan atau pelonggaran dalam waktu dekat.

## **Data Cleaning**

```python
print("Starting Data Cleaning Process...")

cols_anomali = ['New Cases', 'New Deaths', 'New Recovered', 'New Active Cases']

for col in cols_anomali:
    if col in df.columns:
        df[col] = df[col].clip(lower=0)

print(f"  - Corrected negative values in columns: {cols_anomali}")

cols_zero_fill = [
    'Total Cases', 'Total Deaths', 'Total Recovered', 'Total Active Cases',
    'New Cases', 'New Deaths', 'New Recovered'
]

cols_existing_fill = [c for c in cols_zero_fill if c in df.columns]
df[cols_existing_fill] = df[cols_existing_fill].fillna(0)

print(f"  - Imputed missing values with 0 for {len(cols_existing_fill)} columns.")

df_provinces = df[df['Location Level'] == 'Province']

df_latest_clean = df_provinces.sort_values('Date').groupby('Location').tail(1)

cols_mandatory = [
    'Total Deaths per Million',
    'Population Density',
    'Total Regencies',
    'Total Cities'
]
df_latest_clean = df_latest_clean.dropna(subset=cols_mandatory)

print(f"  - Snapshot Dataset Prepared (Provinces): {len(df_latest_clean)} provinces retained.")

df_national_clean = df[df['Location'] == 'Indonesia'].sort_values('Date').set_index('Date')

df_national_clean = df_national_clean.resample('D').mean(numeric_only=True)

if 'New Cases' in df_national_clean.columns:
    df_national_clean['New Cases'] = df_national_clean['New Cases'].interpolate(method='linear')

df_national_clean = df_national_clean.reset_index()

print(f"  - Time Series Dataset Prepared (National): {len(df_national_clean)} days processed.")

print("\nData Cleaning Completed Successfully.")
print(f"  - df_latest_clean shape (Provinces): {df_latest_clean.shape}")
print(f"  - df_national_clean shape (National): {df_national_clean.shape}")

print("\nInfo Dataset Setelah Cleaning:")
print("DATA SNAPSHOT (PROVINCES):")
df_latest_clean.info()
print("\nDATA TIME SERIES (NATIONAL):")
df_national_clean.info()
```

```python
print("Starting Data Preprocessing...")

# Preprocessing untuk DBSCAN
# Tujuan: Segmentasi Provinsi berdasarkan tingkat keparahan & urbanisasi
# Data: df_latest_clean
print("\n[1] Preprocessing for Clustering (DBSCAN)...")

features_clust = [
    'Total Cases per Million',
    'Total Deaths per Million',
    'Population Density',
    'Total Cities'
]

X_clust = df_latest_clean[features_clust].copy()

scaler_clust = StandardScaler()
X_clust_scaled = scaler_clust.fit_transform(X_clust)

print(f"  - Features selected: {features_clust}")
print(f"  - Scaling applied (StandardScaler). Shape: {X_clust_scaled.shape}")


# Preprocessing untuk K-Means
# Tujuan: Segmentasi Provinsi berdasarkan karakteristik wilayah
# Data: df_latest_clean
print("\n[2] Preprocessing for Clustering (K-Means)...")

# Seleksi fitur numerik untuk clustering
features_kmeans = [
    'Population Density',
    'Total Regencies',
    'Total Cities',
    'Total Districts',
    'Area (km2)'
]

# Ambil data
X_kmeans = df_latest_clean[features_kmeans].copy()

# Scaling (StandardScaler)
scaler_kmeans = StandardScaler()
X_kmeans_scaled = scaler_kmeans.fit_transform(X_kmeans)

print(f"  - Features selected: {features_kmeans}")
print(f"  - Scaling applied (StandardScaler). Shape: {X_kmeans_scaled.shape}")


# Preprocessing untuk MLP
# Tujuan: Forecasting 'New Cases' (Nasional)
# Data: df_national_clean
print("\n[3] Preprocessing for Time Series (MLP)...")

ts_data = df_national_clean[['Date', 'New Cases']].copy()

# Feature Engineering: Lag Features
lags = [1, 7, 14, 30]
for lag in lags:
    ts_data[f'Lag_{lag}'] = ts_data['New Cases'].shift(lag)

# Hapus NaN akibat shifting
ts_data = ts_data.dropna()

# Definisi X dan y
X_ts = ts_data[[f'Lag_{lag}' for lag in lags]]
y_ts = ts_data['New Cases']

# Train-Test Split (Sequential)
train_size = int(len(X_ts) * 0.8)
X_train_ts, X_test_ts = X_ts.iloc[:train_size], X_ts.iloc[train_size:]
y_train_ts, y_test_ts = y_ts.iloc[:train_size], y_ts.iloc[train_size:]

# Scaling (MinMaxScaler)
scaler_ts = MinMaxScaler()
X_train_ts_scaled = scaler_ts.fit_transform(X_train_ts)
X_test_ts_scaled = scaler_ts.transform(X_test_ts)

print(f"  - Lag Features created: {lags}")
print(f"  - Train Data: {X_train_ts_scaled.shape}, Test Data: {X_test_ts_scaled.shape}")
print(f"  - Scaling applied (MinMaxScaler).")

print("\nData Preprocessing Completed Successfully.")
```

## **DBSCAN**

```python
print("Starting DBSCAN Clustering with Grid Search...")

features_clust = [
    'Total Cases per Million',
    'Total Deaths per Million',
    'Population Density',
    'Total Cities'
]

X_clust = df_latest_clean[features_clust].copy()

scaler = StandardScaler()
X_clust_scaled = scaler.fit_transform(X_clust)

# Grid Search
eps_values = np.arange(0.3, 1.6, 0.1)
min_samples_values = range(3, 8)

results = []

for eps in eps_values:
    for min_s in min_samples_values:

        db = DBSCAN(eps=eps, min_samples=min_s)
        labels = db.fit_predict(X_clust_scaled)

        unique_labels = set(labels)
        n_clusters = len(unique_labels - {-1})
        n_noise = list(labels).count(-1)

        if n_clusters >= 2:
            mask = labels != -1
            sil = silhouette_score(
                X_clust_scaled[mask],
                labels[mask]
            )
        else:
            sil = np.nan

        results.append({
            "eps": eps,
            "min_samples": min_s,
            "n_clusters": n_clusters,
            "n_noise": n_noise,
            "silhouette": sil
        })

df_dbscan_results = pd.DataFrame(results)

print("\nTop 10 Grid Search Results:")
print(
    df_dbscan_results
    .sort_values(by="silhouette", ascending=False)
    .head(10)
)

best_row = (
    df_dbscan_results
    .dropna()
    .query("n_clusters >= 2")
    .sort_values(
        by=["silhouette", "n_noise"],
        ascending=[False, True]
    )
    .iloc[0]
)

best_eps = best_row["eps"]
best_min_samples = int(best_row["min_samples"])

print("\nBest DBSCAN Parameters:")
print(f"eps         : {best_eps}")
print(f"min_samples : {best_min_samples}")
print(f"clusters    : {best_row['n_clusters']}")
print(f"noise pts   : {best_row['n_noise']}")
print(f"silhouette  : {best_row['silhouette']:.3f}")

dbscan_best = DBSCAN(eps=best_eps, min_samples=best_min_samples)
df_latest_clean["Cluster_DBSCAN"] = dbscan_best.fit_predict(X_clust_scaled)

plot_df = df_latest_clean.copy()
plot_df['Cluster_Label'] = plot_df['Cluster_DBSCAN'].apply(
    lambda x: 'Outlier' if x == -1 else f'Cluster {x}'
)
plot_df = plot_df.sort_values('Cluster_Label')

plt.figure(figsize=(12, 8))

unique_clusters = plot_df['Cluster_Label'].unique()
palette = sns.color_palette("bright", len(unique_clusters))
color_dict = dict(zip(unique_clusters, palette))

if 'Outlier' in color_dict:
    color_dict['Outlier'] = '#333333'

sns.scatterplot(
    data=plot_df,
    x='Population Density',
    y='Total Deaths per Million',
    hue='Cluster_Label',
    style='Cluster_Label',
    palette=color_dict,
    s=150,
    alpha=0.75,
    edgecolor='white',
    linewidth=1
)

plt.xscale('log')

noise_data = plot_df[plot_df['Cluster_Label'] == 'Outlier']
for idx, row in noise_data.iterrows():
    plt.text(
        row['Population Density'],
        row['Total Deaths per Million'],
        row['Location'],
        fontsize=9,
        color='black',
        fontweight='bold',
        ha='right',
        va='bottom'
    )

plt.title(f"DBSCAN Clustering",
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Population Density (People/km²) - Log Scale", fontsize=12)
plt.ylabel("Total Deaths per Million", fontsize=12)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', title='Classification', frameon=True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))
counts = plot_df['Cluster_Label'].value_counts().sort_index()
sns.barplot(x=counts.index, y=counts.values, palette='tab10')
plt.title("Number of Provinces per DBSCAN Cluster")
plt.xlabel("Cluster")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("DBSCAN CLUSTER MEMBERS")

for label in sorted(df_latest_clean['Cluster_DBSCAN'].unique()):
    cluster_name = "OUTLIER" if label == -1 else f"CLUSTER {label}"
    members = df_latest_clean[
        df_latest_clean['Cluster_DBSCAN'] == label
    ]['Location'].tolist()

    print(f"\n>> {cluster_name} ({len(members)} Provinces)")
    print(", ".join(members))
```

## **K-Means**

```python
print("Starting K-Means Clustering...")

features_clust = [
    'Total Cases per Million',
    'Total Deaths per Million',
    'Population Density',
    'Total Cities'
]

X_clust = df_latest_clean[features_clust].copy()

scaler_clust = StandardScaler()
X_clust_scaled = scaler_clust.fit_transform(X_clust)

silhouette_scores = []
K_range = range(3, 11)   # K minimal = 3

print("\n" + "=" * 45)
print(f"{'K (Clusters)':<15} | {'Silhouette Score'}")
print("=" * 45)

for k in K_range:
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    labels = kmeans.fit_predict(X_clust_scaled)
    score = silhouette_score(X_clust_scaled, labels)
    silhouette_scores.append(score)

    print(f"{k:<15} | {score:.4f}")

print("=" * 45)

# Pilih K terbaik berdasarkan silhouette score
best_k = K_range[silhouette_scores.index(max(silhouette_scores))]
print(f"\nBest K selected based on Silhouette Score (K ≥ 3): K={best_k}")

inertia = []

for k in K_range:
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )
    kmeans.fit(X_clust_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(7, 5))
plt.plot(
    list(K_range),
    inertia,
    marker='o',
    linestyle='-',
    linewidth=2
)
plt.title('Elbow Method (K ≥ 3)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.xticks(list(K_range))
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

selected_k = best_k
print(f"\nExecuting K-Means with K={selected_k}...")

kmeans_final = KMeans(
    n_clusters=selected_k,
    random_state=42,
    n_init=10
)

df_latest_clean['Cluster_KMeans'] = kmeans_final.fit_predict(X_clust_scaled)

plt.figure(figsize=(12, 7))
sns.scatterplot(
    data=df_latest_clean,
    x='Population Density',
    y='Total Deaths per Million',
    hue='Cluster_KMeans',
    palette='tab10',
    s=150,
    alpha=0.9,
    edgecolor=None
)

for _, row in df_latest_clean.iterrows():
    if row['Population Density'] > 2000 or row['Total Deaths per Million'] > 1200:
        plt.text(
            row['Population Density'] + 50,
            row['Total Deaths per Million'],
            row['Location'],
            fontsize=9,
            fontweight='bold'
        )

plt.title(f'K-Means Segmentation (K={selected_k})', fontsize=14)
plt.xlabel('Population Density (People/km²)')
plt.ylabel('Total Deaths per Million')
plt.legend(title='Cluster ID')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

cluster_means = (
    df_latest_clean
    .groupby('Cluster_KMeans')[features_clust]
    .mean()
    .sort_index()
)

fig, axes = plt.subplots(1, len(features_clust), figsize=(18, 5))

for i, col in enumerate(features_clust):
    sns.barplot(
        x=cluster_means.index,
        y=cluster_means[col],
        ax=axes[i],
        palette='tab10'
    )
    axes[i].set_title(f'Avg {col}', fontsize=10)
    axes[i].set_xlabel('Cluster ID')
    axes[i].set_ylabel('')

plt.suptitle('Cluster Profiling (Mean Values)', fontsize=14)
plt.tight_layout()
plt.show()

print("\nK-Means Cluster Members")
for label in sorted(df_latest_clean['Cluster_KMeans'].unique()):
    members = df_latest_clean[
        df_latest_clean['Cluster_KMeans'] == label
    ]['Location'].tolist()

    print(f"\n>> CLUSTER {label} ({len(members)} Provinces):")
    print(", ".join(members))

print("\nK-Means Clustering Completed Successfully.")
```

## **MLP**

```python
print("Starting Time Series Forecasting...")

if 'df_national_clean' not in locals():
    df_ts = df[df['Location'] == 'Indonesia'].sort_values('Date').set_index('Date')
    df_ts = df_ts.resample('D').mean(numeric_only=True) # Resample Harian
    df_ts['New Cases'] = df_ts['New Cases'].fillna(0)
else:
    df_ts = df_national_clean.set_index('Date').copy()

series = df_ts['New Cases']

lags = [1, 7, 14, 30]
df_supervised = pd.DataFrame(series)
for lag in lags:
    df_supervised[f'Lag_{lag}'] = series.shift(lag)

df_supervised = df_supervised.dropna()
print(f"Features created: {[f'Lag_{lag}' for lag in lags]}")

train_size = int(len(df_supervised) * 0.8)
train_data = df_supervised.iloc[:train_size]
test_data = df_supervised.iloc[train_size:]

X_train = train_data.drop('New Cases', axis=1)
y_train = train_data['New Cases']
X_test = test_data.drop('New Cases', axis=1)
y_test = test_data['New Cases']

scaler_X = MinMaxScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

scaler_y = MinMaxScaler()
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).ravel()
y_test_scaled = scaler_y.transform(y_test.values.reshape(-1, 1)).ravel()

print("Training MLP Neural Network...")
mlp = MLPRegressor(
    hidden_layer_sizes=(64, 32),
    activation='relu',
    solver='adam',
    max_iter=500,
    random_state=42,
    early_stopping=True
)
mlp.fit(X_train_scaled, y_train_scaled)

y_pred_scaled = mlp.predict(X_test_scaled)

y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).ravel()
y_actual = y_test.values

r2 = r2_score(y_actual, y_pred)

print(f"\nModel Performance")
print(f"R2 Score : {r2:.4f}")

plt.figure(figsize=(14, 7))

plt.plot(test_data.index, y_actual, label='Actual New Cases', color='#2980b9', linewidth=2)
plt.plot(test_data.index, y_pred, label='Predicted (MLP Model)', color='#e74c3c', linestyle='--', linewidth=2)

plt.title('MLP', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('New Cases', fontsize=12)
plt.legend(loc='upper left', fontsize=11)
plt.grid(True, linestyle='--', alpha=0.5)

plt.fill_between(test_data.index, y_actual, y_pred, color='gray', alpha=0.1, label='Prediction Error')

plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 5))
last_n = 90
plt.plot(test_data.index[-last_n:], y_actual[-last_n:], label='Actual', color='#2980b9', marker='o', markersize=4)
plt.plot(test_data.index[-last_n:], y_pred[-last_n:], label='Predicted', color='#e74c3c', linestyle='--', marker='x', markersize=4)
plt.title(f'Zoomed In: Last {last_n} Days Performance', fontsize=14)
plt.grid(True, alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
